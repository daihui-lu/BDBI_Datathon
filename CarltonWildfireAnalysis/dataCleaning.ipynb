```python
import gdal
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
def read_dataset(SUBDATASET_NAME, FILEPATH):
    dataset = gdal.Open(FILEPATH)
    path = ''
    for sub, description in dataset.GetSubDatasets():
        if (description.endswith(SUBDATASET_NAME)):
            path = sub
            break
    if(path == ''):
        print(SUBDATASET_NAME + ' not found')
        return
    subdataset = gdal.Open(path)
    subdataset = subdataset.ReadAsArray()
    subdataset = pd.DataFrame(subdataset)
    return subdataset
```

# Precipitation

```python
fileNames=os.listdir('./Calton/Precipitation')
```

```python
Preci=pd.DataFrame()
for k in range(0,6,2):
    file_path = './Calton/Precipitation/'+fileNames[k]
    file_path2 = './Calton/Precipitation/'+fileNames[k+1]
    df = read_dataset('IR_Precip_Est_A ascending (32-bit floating-point)',file_path)
    df2 = read_dataset('IR_Precip_Est_A ascending (32-bit floating-point)',file_path2)
    
    data=df.iloc[39:51, 40:88]
    data2=df2.iloc[39:51, 40:88]
    Aver=(data+data2)/2
    
    result=pd.DataFrame(np.zeros((1200, 2400)))
    for i in range(1200):
        if i%100==0:
            print(i)
        for j in range(2400):
            result.iloc[i,j]=Aver.iloc[i//100,j//100]
    
    result=result.stack().reset_index()
    result['time']=i/2
    Preci=Preci.append(result)
```

```python
Preci=Preci.rename(index=str,columns={0: "Preci"})
```

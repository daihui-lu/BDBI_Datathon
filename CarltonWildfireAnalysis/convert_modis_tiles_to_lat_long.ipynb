# Inverse mapping

```python
import pandas as pd
import numpy as np
```

```python
row = 1200
col = 1200
lat = np.zeros([row,col*2])
long = np.zeros([row,col*2])
R = 6371007.181
T = 1111950
xmin = -20015109
ymax = 10007555 
w = T/1200
H = [9, 10]
V = 4

# for h in (H):
for i in range(row):
    y = ymax - V*T - (i+0.5)*w
    for j in range(col*2):
        if j < col:
            h = 9
            jj = j
        else:
            h = 10
            jj = j - col
            
        x = (jj + 0.5)* w + h*T + xmin
        
        lat[i][j] = y/R
        long[i][j] = x/R/np.cos(lat[i][j])        
lat *= (180/np.pi) 
long *= (180/np.pi) 
lat_df = pd.DataFrame(lat)
long_df = pd.DataFrame(long)
long_df.to_csv(r'longitude.csv', index = False)
lat_df.to_csv(r'latitude.csv', index = False)
```
